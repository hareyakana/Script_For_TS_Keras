```python
import numpy as np
import matplotlib.pyplot as plt
from root_pandas import read_root
from sklearn.preprocessing import normalize

plt.rcParams['figure.figsize'] = 5.0, 4.0
file = "Piled_2.root"

pmtall_1 = read_root(file,"tree",columns=["PMTALL"],flatten=["PMTALL"])
l_1,h_1 = pmtall_1.shape
l_1=int(l_1/4480)
psdpara = read_root(file,"tree",columns=["PSDPara"],flatten=["PSDPara"])
l,h = psdpara.shape
l=int(l/34)
psddoublep = read_root(file,"tree",columns=["PSDDoubleP"],flatten=["PSDDoubleP"])
ldp, hdp = psddoublep.shape
ldp = int(ldp/22)
ratior = read_root(file,"tree",columns=["RatioR"],flatten=["RatioR"])
lrr, hrr = ratior.shape
lrr = int(lrr/3)
energy = read_root(file,"tree",columns=["Energy"],flatten=["Energy"])
le,he = energy.shape
le = int(le/3)
ratio4us = read_root(file,"tree",columns=["Ratio4us"])
ratio = read_root(file,"tree",columns=["Ratio"])

quality = read_root(file,"tree",columns=["DataQuality"],flatten=["DataQuality"])
lq,hq = quality.shape
lq = int(lq/4)
def normqual(x,n,z):
    return x.DataQuality.values.reshape(n,z)
DataQuality = normqual(quality,lq,4)
# DataQuality0 = np.expand_dims(DataQuality[:,0],axis=1)
# DataQuality1 = np.expand_dims(DataQuality[:,1],axis=1)
# DataQuality2 = np.expand_dims(DataQuality[:,2],axis=1)
# DataQuality3 = np.expand_dims(DataQuality[:,3],axis=1)

def normalized_PSD(x,n,l):
    y = x.PSDPara.values.reshape(n,l)
    return y

def normalized_PSDDoubleP(x,n,l):
    y = x.PSDDoubleP.values.reshape(n,l)
    return y

def normalized_RatioR(x,n,l):
    y = x.RatioR.values.reshape(n,l)
    return y

def normalized_Energy(x,n,l):
    y = x.Energy.values.reshape(n,l)
    return y

PSDPara = normalized_PSD(psdpara,l,34)
PSDDoubleP = normalized_PSDDoubleP(psddoublep,ldp,22)
RatioR = normalized_RatioR(ratior,lrr,3)
Energy = normalized_Energy(energy,le,3)
Ratio4us = ratio4us.values
Ratio = ratio.values

PSDPara1 = np.expand_dims(PSDPara[:,1],axis=1)
PSDPara7 = np.expand_dims(PSDPara[:,7],axis=1)
PSDPara13 = np.expand_dims(PSDPara[:,13],axis=1)
PSDDoubleP16 = np.expand_dims(PSDDoubleP[:,16],axis=1)
Energy1 = np.expand_dims(Energy[:,1],axis=1)
RatioR1 = np.expand_dims(RatioR[:,1],axis=1)

del PSDPara,PSDDoubleP,psdpara,psddoublep,Energy,RatioR

def NORMPMTALL(pmtall,num_entries):
    x = pmtall.PMTALL.values.reshape(num_entries,4480)
    x_abs = abs(x-15200)
    extra = np.arange(1000,4480)
    x_mod = np.delete(x_abs,extra,axis=1)
    x_nor = normalize(x_mod,norm="l1")
    return x_nor

Pmtall_1 = NORMPMTALL(pmtall_1,l_1)

del pmtall_1

x_1 = np.arange(len(Pmtall_1[1]))

print(len(Pmtall_1))

dummy = 0
for i in range(100):
    k_1 = np.random.randint(len(Pmtall_1))
    plt.figure()
    plt.title(file)
    plt.plot(x_1,Pmtall_1[k_1])
    plt.show()
    print("PSDPara[1],PSDPara[7],PSDPara[13],PSDDoubleP[16],RatioR[1],Ratio4us,Ratio,Energy[1]")
    print(PSDPara1[k_1],PSDPara7[k_1],PSDPara13[k_1],PSDDoubleP16[k_1],RatioR1[k_1],Ratio4us[k_1],Ratio[k_1],Energy1[k_1],DataQuality[k_1])
    
#     if Ratio4us[i]>1 and DataQuality[i,0]==0 and DataQuality[i,1]==0 and DataQuality[i,2]==0 and DataQuality[i,3]==0:
#         plt.figure()
#         plt.title(file)
#         plt.plot(x_1,Pmtall_1[i])
#         plt.show()
#         print("PSDPara[1],PSDPara[7],PSDPara[13],PSDDoubleP[16],RatioR[1],Ratio4us,Ratio,Energy[1]")
#         print(PSDPara1[i],PSDPara7[i],PSDPara13[i],PSDDoubleP16[i],RatioR1[i],Ratio4us[i],Ratio[i],Energy1[i],DataQuality[i])
#         dummy+=1
#     if dummy == 50:
#         break
```
